```python
import pandas as pd
import numpy as np
```

```python
filename = "./starter/data/census.csv"
```

```python
df = pd.read_csv(filename)
df.head()
```

```python
#Check data types, NULL values
df.info()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt

cat_features = df.columns[np.where(df.dtypes == object)[0]].tolist()
for name in cat_features:
    print(df[name].value_counts().plot(kind='bar', title='Distributrion of {0}'.format(name)))
    plt.show()
```

```python
#Removing leading/trailing whitespaces
df.columns = [col.strip() for col in df.columns]
df.info()
```

```python
cat_features = df.columns[np.where(df.dtypes == object)[0]].tolist()
for col in cat_features:
    df[col] = df[col].str.strip()
```

```python
df.head()
```

```python
#save
df.to_csv(filename, index=False)
```
